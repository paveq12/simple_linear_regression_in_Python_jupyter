```python
import numpy as np
import pandas as pd
import sklearn.linear_model as m
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/home/pawel/Pobrane/day.csv', sep = ',')
```

```python
x = data.loc[: , "windspeed"] * 67
```

```python
y =  data.loc[: , "registered"]
```

```python
model = m.LinearRegression()
model.fit(x[:, np.newaxis], y)
```

```python
xfit = np.linspace(-2, 35, 20)
yfit = model.predict(xfit[:, np.newaxis])
```

```python
plt.scatter(x, y,  color='black')
plt.plot(xfit, yfit, color='red', linewidth=3)
plt.show()
```
